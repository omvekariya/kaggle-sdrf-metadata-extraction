# Harmonizing the Data of your Data - EDA

Exploratory Data Analysis for the Kaggle competition.

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

from src.data.pdf_parser import PDFParser
from src.data.xml_parser import XMLParser
from src.data.dataset import CitationDataset

plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline
```

## 1. Load Data

```python
# Set paths
DATA_DIR = Path('../data/raw')

# List available files
pdf_files = list(DATA_DIR.glob('**/*.pdf'))
xml_files = list(DATA_DIR.glob('**/*.xml'))

print(f'PDF files: {len(pdf_files)}')
print(f'XML files: {len(xml_files)}')
```

## 2. Explore PDF Files

```python
# Initialize parser
pdf_parser = PDFParser()

# Parse a sample PDF (update path after downloading data)
if pdf_files:
    sample_pdf = pdf_files[0]
    text = pdf_parser.extract_text(sample_pdf)
    print(f'Sample PDF: {sample_pdf.name}')
    print(f'Text length: {len(text)} characters')
    print(f'\nFirst 1000 characters:\n{text[:1000]}')
```

## 3. Explore XML Files

```python
# Initialize parser
xml_parser = XMLParser()

# Parse a sample XML (update path after downloading data)
if xml_files:
    sample_xml = xml_files[0]
    sections = xml_parser.extract_sections(sample_xml)
    print(f'Sample XML: {sample_xml.name}')
    print(f'Sections found: {list(sections.keys())}')
```

## 4. Citation Pattern Analysis

```python
from src.models.citation_extractor import CitationExtractor

extractor = CitationExtractor()

# Test extraction on sample text
sample_text = """
The crystal structure was deposited in the Protein Data Bank under accession code 6TAP.
RNA-seq data are available at GEO (GSE123456). We also used previously published
data from DOI: 10.1234/example.2023 which was downloaded from the repository.
"""

identifiers = extractor.extract_identifiers(sample_text)
for ident in identifiers:
    print(f"Type: {ident['type']}, Value: {ident['identifier']}")
```

## 5. Labels Analysis

```python
# Load training labels (update path after downloading data)
labels_path = DATA_DIR / 'train_labels.csv'

if labels_path.exists():
    labels_df = pd.read_csv(labels_path)
    print(labels_df.head())
    print(f'\nShape: {labels_df.shape}')
    print(f'\nColumns: {labels_df.columns.tolist()}')
else:
    print('Labels file not found. Download competition data first.')
```
